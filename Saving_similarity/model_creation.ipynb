```python
dd = pd.read_csv('./data_OrangeNov2019.csv',sep='|')
def renameCols (collist):
    dnewcol={}
    for c in collist:
        newname = c.split(';')[0].replace(' ','')
        dnewcol[c]=newname
    return dnewcol
dd.rename(columns=renameCols(dd.columns),inplace=True)
```
